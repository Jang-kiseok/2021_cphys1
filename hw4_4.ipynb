```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#(Differentiation) Trigonometric Function
#3.1 sinx finite difference
xdata = np.arange(0,np.pi*2,0.01)
def f(x):
    return np.sin(x)

plt.xlim(0,np.pi*2);plt.ylim(-2,2)
plt.plot(xdata,f(xdata));
```

```python
def trapez(f,a,b,n):
    s, x, dx = 0, a, (b-a)/n
    while(x<b):
        s += dx*(0.5*f(x)+0.5*f(x+dx))
        x = x + dx
    return s

def for_diff(f,x,dx):
    return (f(x+dx)-f(x))/dx
def back_diff(f,x,dx):
    return (f(x)-f(x-dx))/dx
def cen_diff(f,x,dx):
    return (f(x+dx/2)-f(x-dx/2))/dx
```

```python
dx=0.1
xdata=np.arange(0,np.pi*2,0.01)
diff1=[for_diff(f,i,dx) for i in xdata]
diff2=[back_diff(f,i,dx) for i in xdata]
diff3=[cen_diff(f,i,dx) for i in xdata]
```

```python
plt.plot(xdata,diff1,label= "$for_diff$")
plt.plot(xdata,diff2,label = "$back_diff$")
plt.plot(xdata,diff3,label = "$cen_diff$")
plt.plot(xdata,f(xdata),label ="$cos{x}$")
plt.grid()
plt.legend();
```
