```python
#라이프니츠 파이공식 그래프
import matplotlib.pyplot as plt
import numpy as np
import math
pi= math.pi

def F(n):
    i,s = 1,0
    S = []
    while(2*i-1<=2*n-1):
        s = s+(-1)**(i+1)/(2*i-1)
        i += 1
        S += [s] 
    return S
S = F(100)
plt.plot(S,label="$Leibniz formula forπ$")

a = [1/4*pi]
plt.plot([0,20,40,60,80,100],[a,a,a,a,a,a],label="$π/4$")
plt.grid()
plt.legend();
```
