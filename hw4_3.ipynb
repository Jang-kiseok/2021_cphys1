```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import integrate
```

```python
#scipy.integrate 
#1)sin(x)
def f(x):
    return np.sin(x)
integrate.quad(f,0,np.pi)
```

```python
xdata = np.arange(0,np.pi*2,0.01)
def f(x):
    return np.sin(x)

plt.xlim(0,np.pi*2);plt.ylim(-2,2)
plt.plot(xdata,f(xdata));
```

```python
#2)cos(x)
def f(x):
    return np.cos(x)
integrate.quad(f,0,np.pi)
```
