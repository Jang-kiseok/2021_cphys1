```python
import random
import numpy as np
```

```python
#차원의구 
#몬테카를로 방법을 이용한 3, 4, 5차원에서 $Cn의 근사값
#2차원
def s_circle(n):
    count=0
    for i in range(n):
        x,y=random.random(), random.random()
        if x*x+y*y<1:
            count+=1
    return 4*count/n
s_circle(100)
```

```python
#3차원
def s_circle(n):
    count=0
    for i in range(n):
        x,y,z=random.random(), random.random(), random.random()
        if x*x+y*y+z*z<1:
            count+=1
    return 8*count/n
s_circle(100)
```

```python
#4차원
def s_circle(n):
    count=0
    for i in range(n):
        x,y,z,w=random.random(), random.random(), random.random(), random.random()
        if x*x+y*y+z*z+w*w<1:
            count+=1
    return 16*count/n
s_circle(100)
```

```python
def s_circle(n):
    count = 0
    for i in range(n):
        x,y,z,w,t = random.random(),random.random(),random.random(),random.random(),random.random()
        if x*x+y*y+z*z+w*w+t*t<1:
            count+=1
    return 32*count/n
s_circle(100)
```

```python
#차원구의 면적근사값
def s_circle(n,t):
    count = 0
    for i in range(n):
        rr = np.zeros(t)
        m = 0
        while(m < t):
            x = random.random()
            rr[m] = x*x
            m+=1
        print(rr,sum(rr))
        if sum(rr)<1:
            count+=1
            
    return 2**t*count/n
s_circle(100,3)
```
