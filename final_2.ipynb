```python
#Bifurcation
from sympy import *
import math
```

```python
t = symbols('t')
r = symbols('r')
x = Function('x')
x,x(t).diff(t)
```

```python
deq = Eq(x(t).diff(t),r*x(t)-x(t)*x(t))
deq
```

```python
#r= -1, x(0)=2
deq = Eq(x(t).diff(t),-1*x(t)-x(t)*x(t))
deq
psol1 = dsolve( deq, ics= {x(0):2} )
psol1 
```

```python
#r= -1, x(0)=-2
deq = Eq(x(t).diff(t),-1*x(t)-x(t)*x(t))
deq
psol2 = dsolve( deq, ics= {x(0):-2} )
psol2
```

```python
#r= 1, x(0)=2
deq = Eq(x(t).diff(t),1*x(t)-x(t)*x(t))
deq
psol3 = dsolve( deq, ics= {x(0):2} )
psol3
```

```python
#r= 1, x(0)=-2
deq = Eq(x(t).diff(t),1*x(t)-x(t)*x(t))
deq
psol4 = dsolve( deq, ics= {x(0):-2} )
psol4
```

```python
plot( psol1.rhs, (t,-1,1))
plot( psol2.rhs, (t,-1,1))
plot( psol3.rhs, (t,-1,1))
plot( psol4.rhs, (t,-1,1))
```
