```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x = [0]  
def random1_D(t):
    M = [1, -1]
    P = [0.5, 0.5]
    for i in range(1, t + 1):
        y = np.random.choice(M, 1, P )
        x.append(x[i-1] + y)  
    return x
t = 100   
T = [i for i in range(t+1)] 

plt.plot(T, random1_D(t))
plt.grid(axis = 'y')
plt.xlabel('$time(s)$')
plt.ylabel('$Position$');

print(x[t])
```
