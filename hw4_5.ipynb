```python
#scipy.misc.derivative 
#3.(1) sin(x)
import matplotlib.pyplot as plt
import numpy as np
from scipy.misc import derivative
derivative(np.sin,np.pi,dx=0.01)
```

```python
xdata=np.arange(0,8,0.01)
plt.plot(xdata,[derivative(np.sin,i,dx=0.01) for i in xdata])
plt.plot(xdata,[derivative(np.sin,i,dx=0.01,n=2) for i in xdata])
```

```python
#3.(2) cos(x)
import matplotlib.pyplot as plt
import numpy as np
from scipy.misc import derivative
derivative(np.cos,np.pi/2,dx=0.01)
```

```python
xdata=np.arange(0,8,0.01)
plt.plot(xdata,[derivative(np.cos,i,dx=0.01) for i in xdata])
plt.plot(xdata,[derivative(np.cos,i,dx=0.01,n=2) for i in xdata])
```
