```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import integrate
```

```python
#2. Trigonometric Function 
#1) sin(x) 함수
def sin(x):
    return np.sin(x)

def trapez(f,a,b,n):
    s, x, dx = 0, a, (b-a)/n
    while(x<b):
        s += dx*(0.5*f(x)+0.5*f(x+dx))
        x = x + dx
    return s
trapez(sin,0,np.pi,100)
```

```python
xdata = np.arange(0,8,0.01)
def Sin(x):
    return np.sin(x)

plt.xlim(0,np.pi*2);plt.ylim(-2,2)
plt.plot(xdata,Sin(xdata));
```

```python
#2) cos(x) 함수
def cos(x):
    return np.cos(x)

def trapez(f,a,b,n):
    s, x, dx = 0, a, (b-a)/n
    while(x<b):
        s += dx*(0.5*f(x)+0.5*f(x+dx))
        x = x + dx
    return s

trapez(cos,0,np.pi*0.5,100)
```

```python
xdata = np.arange(0,8,0.01)
def Cos(x):
    return np.cos(x)

plt.xlim(0,np.pi*2);plt.ylim(-2,2)
plt.plot(xdata,Cos(xdata));
```
