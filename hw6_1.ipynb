```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint
```

```python
#1. pendulum
#W 값 변환
#W=0
state0 = [0.1,0] 
t = np.arange(0,100,0.1)
def f(state, t):
    θ, w = state
    return w, -gL*np.sin(θ)

gL = 1
states = odeint(f, state0, t)
plt.plot(t,states[:,0],label = 'W = 0')
plt.legend();

#W=1
state0 = [0.1,1] 
t = np.arange(0,100,0.1)
def f(state, t):
    θ, w = state
    return w, -gL*np.sin(θ)

gL = 1
states = odeint(f, state0, t)
plt.plot(t,states[:,0],label = 'W = 1')
plt.legend();

#W=1.9
state0 = [0.1,1.9] 
t = np.arange(0,100,0.1)
def f(state, t):
    θ, w = state
    return w, -gL*np.sin(θ)

gL = 1
states = odeint(f, state0, t)
plt.plot(t,states[:,0],label = 'W = 1.9')
plt.legend();
```

```python
#3. Damped Pendulum
w = 1
b = 0.1
state0 = [1.0,0.0] # (x0 , v0)
t = np.arange(0.0,50.0,0.01)
def f(state,t):
    x,v = state
    return v, -2*b*v-w**2*x
states = odeint(f,state0,t)
plt.plot(states[:,0])
```

```python
#4. Forced Pendulum
#V값 변환
#V = 5
state0 = [1,0]  # [theta0, omega0]
β = 0.1
gL = 1
A = 2
v = 5
t = np.arange(0,100,0.1)
def f(state, t):
    θ, w = state
    return w, -β*w-gL*np.sin(θ)+A*np.cos(v*t)
states = odeint(f, state0, t)
plt.plot(t,states[:,0],label = 'v = 5')
plt.legend()

#V = 30
state0 = [1,0]  # [theta0, omega0]
β = 0.1
gL = 1
A = 2
v = 30
t = np.arange(0,100,0.1)
def f(state, t):
    θ, w = state
    return w, -β*w-gL*np.sin(θ)+A*np.cos(v*t)
states = odeint(f, state0, t)
plt.plot(t,states[:,0],label = 'v = 30')
plt.legend()

#V = 50
state0 = [1,0]  # [theta0, omega0]
β = 0.1
gL = 1
A = 2
v = 50
t = np.arange(0,100,0.1)
def f(state, t):
    θ, w = state
    return w, -β*w-gL*np.sin(θ)+A*np.cos(v*t)
states = odeint(f, state0, t)
plt.plot(t,states[:,0],label = 'v = 30')
plt.legend()
```
